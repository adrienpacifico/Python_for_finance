#### Lire des tableaux sur des pages web

Si on a envie de récupérer de la donnée du web, on peut web-scraper de la donnée. Cela demande généralement d'apprendre à maitriser une ou plusieurs librairie de scaping tel que beautifullsoup, scrapy (ou selenium). Cela vous demande en plus de connaitre un peu de HTML.

Cependant si dans une page web, l'information d'un tableau est représenté sous forme de "\<tab>" on peut directement récupérer l'information avec pandas.

```python
pd.read_html("https://www.euribor-rates.eu/fr/taux-euribor-actuels/")[0].T
```

```python
df = pd.read_html("https://www.euribor-rates.eu/fr/taux-euribor-actuels/")[0]
```

```python
df
```

```python
df = df.applymap(lambda x: x[:-2].replace(",","")).set_index("Unnamed: 0").astype("float")/1000/100
```

#### Enregistrer de la donnée

Pour enregistrer de la donnée, il existe plusieurs solutions en fonction de son usage.

Ce que je recommande sont:
- csv: super interopérable, mais très très lent
- parquet: très mature, permet la compression, rapide
- feather: utilise appache arrow, très rapide, besoin de bien formater la donnée en amont, pas de compression

Je pense qu'utiliser hdfs (hadoop file system), n'est plus trop avantageux aujourd'hui.
